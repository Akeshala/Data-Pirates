```python
%matplotlib notebook
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("new_train_data_akeshala_nipun_samal.csv", header=0)
df
```

```python
y=df['Reservation_Status']

a=['Gender', 'Age', 'Ethnicity','Educational_Level', 'Income ', 'Country_region', 'Hotel_Type','No_of_Expected_Days_to_stay', 'Checking_Month', 'no_of_days_from_booking_to_ckeckin', 'Adults', 'Children','Babies', 'Meal_Type', 'Visted_Previously ', 'Previous_Cancellations','Deposit_type','Booking_channel','Required_Car_Parking','Use_Promotion','Discount_Rate','Room_Rate']

```

```python
X=df[['Gender', 'Age', 'Ethnicity','Educational_Level', 'Income', 'Country_region', 'Hotel_Type','No_of_Expected_Days_to_stay', 'Checking_Month', 'no_of_days_from_booking_to_ckeckin', 'Adults', 'Children','Babies', 'Meal_Type', 'Visted_Previously', 'Previous_Cancellations','Deposit_type','Booking_channel','Required_Car_Parking','Use_Promotion','Discount_Rate','Room_Rate']]
```

```python
from sklearn import preprocessing
X = preprocessing.scale(X)
```

```python
x_train, x_validation, y_train, y_validation = train_test_split(X, y, test_size = 0.2)
```

```python
# Instantiate
from sklearn.ensemble import AdaBoostClassifier
abc = AdaBoostClassifier(algorithm='SAMME.R', base_estimator=None,
        learning_rate=1.8, n_estimators=5000, random_state=0)

# Fit
abc.fit(x_train, y_train)
```

```python
y_pred = abc.predict(x_validation)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_validation,y_pred))

print(classification_report(y_validation,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=100,max_features='auto',random_state=0,
                          max_depth=10)
clf.fit(x_train,y_train)




y_pred=clf.predict(x_validation)
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_validation, y_pred))
```
